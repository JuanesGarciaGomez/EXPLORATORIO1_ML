## Comparacion tipo de tarjeta verificando si es fraude o no

```python
import matplotlib.pyplot as plt

fraud_by_type = df.groupby("Card_Type")["Fraud_Label"].apply(
    lambda x: (x == "Fraud").sum() / len(x) * 100
).reset_index()
fraud_by_type.columns = ["Card_Type", "Fraud_Percentage"]

plt.figure(figsize=(8, 5))
bars = plt.bar(fraud_by_type["Card_Type"], fraud_by_type["Fraud_Percentage"], color=["#2E75B6", "#1F4E79", "#4BACC6"])

for bar, val in zip(bars, fraud_by_type["Fraud_Percentage"]):
    plt.text(bar.get_x() + bar.get_width()/2, bar.get_height() + 0.1, f"{val:.1f}%", ha="center", fontsize=11)

plt.title("Porcentaje de Fraude por Tipo de tarjeta", fontsize=13)
plt.xlabel("Tipo de tarjeta")
plt.ylabel("% Fraude")
plt.tight_layout()
plt.show()
```

Se puede observar a partir de la grafica anterior que el porcentaje de fraude a partir de las tarjetas de credito o debito es muy similar tentiendo un 4.8% y 4.9% respectivamente.

## Tasa de fraude según horario inusual de transacción

```python
import matplotlib.pyplot as plt

fraud_by_unusual = df.groupby("Unusual_Time_Transaction")["Fraud_Label"].apply(
    lambda x: (x == "Fraud").sum() / len(x) * 100
).reset_index()
fraud_by_unusual.columns = ["Unusual_Time_Transaction", "Fraud_Percentage"]

plt.figure(figsize=(7, 5))
bars = plt.barh(fraud_by_unusual["Unusual_Time_Transaction"], 
                fraud_by_unusual["Fraud_Percentage"], 
                color=["#2E75B6", "#C00000"])

for bar, val in zip(bars, fraud_by_unusual["Fraud_Percentage"]):
    plt.text(bar.get_width() + 0.1, bar.get_y() + bar.get_height()/2, f"{val:.1f}%", va="center", fontsize=11)

plt.title("Porcentaje de Fraude por Transacción en Horario Inusual", fontsize=13)
plt.ylabel("Horario Inusual")
plt.xlabel("% Fraude")
plt.yticks(ticks=[0, 1], labels=["No", "Yes"])
plt.tight_layout()
plt.show()
```

La gráfica revela que las transacciones realizadas en horario inusual presentan una tasa de fraude del 6.1%, casi el doble frente al 3.6% registrado en horario normal, lo que sugiere que el horario es una variable a tener en cuenta para la deteccion de fraudes.

```python
import matplotlib.pyplot as plt

fraud_by_intl = df.groupby("Is_International_Transaction")["Fraud_Label"].apply(
    lambda x: (x == "Fraud").sum() / len(x) * 100
).reset_index()
fraud_by_intl.columns = ["Is_International_Transaction", "Fraud_Percentage"]

plt.figure(figsize=(7, 5))
bars = plt.barh(fraud_by_intl["Is_International_Transaction"],
                fraud_by_intl["Fraud_Percentage"],
                color=["#2E75B6", "#C00000"])

for bar, val in zip(bars, fraud_by_intl["Fraud_Percentage"]):
    plt.text(bar.get_width() + 0.1, bar.get_y() + bar.get_height()/2, f"{val:.1f}%", va="center", fontsize=11)

plt.title("Tasa de fraude según transacción internacional", fontsize=13)
plt.ylabel("Transacción Internacional")
plt.xlabel("% Fraude")
plt.yticks(ticks=[0, 1], labels=["No", "Yes"])
plt.tight_layout()
plt.show()
```

Las transacciones internacionales registran una tasa de fraude del 6.1%, prácticamente el doble respecto al 3.5% observado en transacciones nacionales, lo que indica que  al ser  internacional  la operación, puede ser un riesgo de transaccion fraudulenta.
